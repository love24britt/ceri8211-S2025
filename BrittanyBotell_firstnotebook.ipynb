```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#varibles
A = np.random.randint(-100, 100, (10000,10000))
x = np.random.randint(-100, 100, (10000,1))
```

```python
#equations
b = np.dot(A,x)
```

```python
#Scatter plot
plt.title("Random Matrix Multiplication")
plt.ylabel("b vector")
plt.xlabel("x vector")
plt.scatter(x,b)
plt.show()
plt.close()
```
